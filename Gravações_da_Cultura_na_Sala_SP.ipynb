```python
mes = 'SETEMBRO' #@param ['JANEIRO', 'FEVEREIRO', 'MARÇO', 'ABRIL', 'MAIO', 'JUNHO', 'JULHO', 'AGOSTO', 'SETEMBRO', 'OUTUBRO', 'NOVEMBRO', 'DEZEMBRO']{allow-input: true}
ano = '2022' #@param [2022, 2023, 2024, 2025, 2026, 2027, 2028, 2029, 2030]{allow-input: true}
```

```python
input = mes
```

```python
import pandas as pd
```

```python
planilha = pd.read_excel('/content/drive/MyDrive/osesp/Controle_de_Gavacoes_2022.xlsx', index_col = 0, header = 2, sheet_name = 'GRAVAÇÕES 2018')
# planilha = planilha.iloc[4:]
planilha.dropna(axis = 0, how = 'all', inplace = True)

planilha
```

```python
planilha.rename(columns = {'CULTURA TUTU': 'CULTURA'}, inplace = True)

mesInput = planilha.loc[planilha['PASTA MENSAL'] == input]

mes = mesInput
mes['CULTURA'] = mes['CULTURA'].str.strip()

mes.dropna(axis = 0, inplace = True, how = 'any', subset = ['CULTURA'])
cultura = mes[(mes['CULTURA'] != '-')]

cultura = cultura[['ORQUESTRA', 'PROJETO', 'PROGRAMA', 'CULTURA']]

cultura['PROGRAMA'].fillna('', inplace = True)
cultura.replace({'PROGRAMA':{'ESP': ''}, 
                 'CULTURA':{'RADIO': 'Rádio', 
                            'RÁDIO': 'Rádio', 
                            'TV': 'Tv', 
                            'RÁDIO & TV': 'Rádio e TV', 
                            'RADIO & TV': 'Rádio e TV', 
                            'RÁDIO E TV': 'Rádio e TV', 
                            'TV E RADIO': 'Rádio e TV', 
                            'TV E RÁDIO': 'Rádio e TV'}}, inplace = True)
cultura
```

```python
cultura['CULTURA'].value_counts()
```

```python
cultura.index = pd.to_datetime(cultura.index, errors = 'coerce').strftime('%d/%b/%Y')
cultura.index = cultura.index.fillna('Erro de data!')
cultura
```

```python
cultura.to_csv(f'/content/drive/MyDrive/downloads/Gravações de {input} de {ano}.csv')
```
